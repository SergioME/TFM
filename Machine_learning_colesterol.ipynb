```python
df[['BMI']] = df[['BMI']].astype(float)
```

```python
a = "Insuficiencia_Ponderal"
b = "Normal"
c = "Sobrepeso"
d = "Obesidad_I"
e = "Obesidad_II"
```

```python
tipos = []
```

```python
for x in df["BMI"]:
    
    if x>=0 and x <=18.4:
        tipos.append(a)
    elif x>=18.5 and x <=24.9:
        tipos.append(b)
    elif x >= 25 and x<=29.9:
        tipos.append(c)
    elif x>=30 and x <=34.9:
        tipos.append(d)
    elif x>=34.9 and x <=39.9:
        tipos.append(e)
```

```python
df["Estado"]=pd.Series(tipos)
```
