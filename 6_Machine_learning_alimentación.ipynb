```python
df.head()
```

```python
df.loc[:,"Sobrepeso"] = 0
df.head()
```

```python
df["Sobrepeso"] = [1 if x>=25 else 0 for x in df["BMI"]]
```

```python
df.sample(8)
```

```python
df.loc[:,"Estado"] = 0
df.head()
```

```python
df.sample(5)
```

```python
df[['BMI']] = df[['BMI']].astype(float)
```

```python
a = "Insuficiencia_Ponderal"
b = "Normal"
c = "Sobrepeso"
d = "Obesidad_I"
e = "Obesidad_II"
```

```python
tipos = []
```

```python
for x in df["BMI"]:
    
    if x>=0 and x <=18.4:
        tipos.append(a)
    elif x>=18.5 and x <=24.9:
        tipos.append(b)
    elif x >= 25 and x<=29.9:
        tipos.append(c)
    elif x>=30 and x <=34.9:
        tipos.append(d)
    elif x>=34.9 and x <=39.9:
        tipos.append(e)
```

```python
df["Estado"]=pd.Series(tipos)
```

```python
##Los ultimos 28 lineas de tabla siempre sale nulo, no he dado con ello, los voy a eliminar
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df[["Fruit","Vegetable","Meat"]] = df[["Fruit","Vegetable","Meat"]].astype(float)
```

```python
def contador_estado (lista):
    Insuficiencia_Ponderal = 0
    Normal = 0
    Sobrepeso = 0
    Obesidad_I = 0
    Obesidad_II = 0
    
    for x in lista:
    
        if x== "Insuficiencia_Ponderal":
            Insuficiencia_Ponderal += +1
        elif x=="Normal":
            Normal += 1
        elif x=="Sobrepeso":
            Sobrepeso += 1
        elif x == "Obesidad_I":
            Obesidad_I += 1
        elif x == "Obesidad_II":
            Obesidad_II += 1
        
    print("Insuficiencia_Ponderal",Insuficiencia_Ponderal)
    print("Normal" , Normal)
    print("Sobrepeso", Sobrepeso)
    print("Obesidad_I", Obesidad_I)
    print("Obesidad_II", Obesidad_II)
```

```python
lista_estado = df["Estado"].tolist()
```

```python
contador_estado(lista_estado)
```

## Train/Test BMI

```python
X = df[["Fruit","Vegetable","Meat"]].values
y = df["Estado"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=25)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rand = RandomForestClassifier(n_estimators=100,max_depth=4)
```

```python
rand.fit(X_train,y_train)
```

```python
from IPython.display import Image  
from sklearn import tree
import pydotplus
```

```python
dot_data = tree.export_graphviz(rand.estimators_[5], out_file=None, 
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png()) 
```

```python
from sklearn.model_selection import cross_val_score
```

```python
print (cross_val_score(rand, X, y, cv=10,scoring="accuracy").mean())
```

## K nearest neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
clf2 = KNeighborsClassifier(n_neighbors=6)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf2.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,clf2.predict(X_test))
```

```python
clf2 = GridSearchCV(estimator=KNeighborsClassifier(n_neighbors=6),
                    param_grid={"n_neighbors":np.arange(1,50)},)
clf2.fit(X,y)
clf2.best_params_
```

```python
clf2.best_score_
```

```python
accuracy_score(y_test,clf2.predict(X_test))
```
